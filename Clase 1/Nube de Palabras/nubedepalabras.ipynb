```python
import os

from os import path
from wordcloud import WordCloud

# Obtenemos el directorio desde el cual estamos ejecutando este notebook
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# Leemos el archivo con los poemas de Neruda
text = open(path.join(d, '20poemas_neruda.txt')).read()

# Generamos una imagen de nube de palabras
wordcloud = WordCloud().generate(text)

# Desplegamos la imagen generada
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# Generamos una segunda imagen, esta vez limitando el tamaño máximo de palabra
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

# Si no funciona matplotlib se puede probar lo siguiente para mostrar la imagen
# image = wordcloud.to_image()
# image.show()
```

### Modificando la nube de palabra
Un problema que tiene esta nube de palabras son las palabras comunes en el idioma español (el, la, los, las, de, etc.). Otro problema es que nos gustaría aumentar su tamaño y modificar el color de fondo.

Para más información, revisar la documentación de la biblioteca wordcloud en https://amueller.github.io/word_cloud/auto_examples/index.html

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
from matplotlib import figure 

# Cargamos las stop words en español
es_stopwords = open("stopwords-es.txt").read().splitlines()

stopwords = set()
stopwords.update(es_stopwords)

# Generate una images de word cloud, esta vez fijamos el color de fondo en blanco
wordcloud = WordCloud(stopwords=stopwords, max_font_size=60, background_color="white").generate(text)

# Desplegamos la imagen usando matplotlib
plt.figure(dpi=150)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
